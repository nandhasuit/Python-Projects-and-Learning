# One hot Encoding

```python
import pandas as pd
import numpy as np
df = pd.read_csv('mercedesbenz.csv',usecols = ['X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
for col in df.columns:
    print(col, ':' ,len(df[col].unique()), 'labels')
```

```python
pd.get_dummies(df,drop_first=True).shape
```

```python
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10 = [i for i in df.X2.value_counts().sort_values(ascending=False).head(20).index]
top_10
```

```python
for lab in top_10:
    df[lab] = np.where(df['X2'] == lab,1,0)
df[['X2']+top_10].head(40)
```

```python
def one_hot(df,value,top_10):
    for lab in top_10:
        df[value+'_'+lab] = np.where(df[value] == lab,1,0)
df = pd.read_csv('mercedesbenz.csv',usecols = ['X1','X2','X3','X4','X5','X6'])
one_hot(df,'X2',top_10)
df.head()
```
